```python
import cv2
import os
import functions
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
number_images=np.zeros(4, dtype=int)
lst = os.listdir("seg_dataset_saranya2021/Green") # your directory path
number_images[0] = len(lst)
lst = os.listdir("seg_dataset_saranya2021/Yellowish_Green") # your directory path
number_images[1] = len(lst)
lst = os.listdir("seg_dataset_saranya2021/Midripen") # your directory path
number_images[2] = len(lst)
lst = os.listdir("seg_dataset_saranya2021/Overripen") # your directory path
number_images[3] = len(lst)
print(number_images)
```

```python
image_indexes=np.zeros(8, dtype=int)
for i in range(image_indexes.shape[0]):
    image_indexes[i]=random.randint(1, number_images[i%4])
```

```python
for i in range(image_indexes.shape[0]):
    if i%4 ==0:
        img = cv2.imread(f"seg_dataset_saranya2021/Green/{str(i)}.jpg")
    if i%4 ==1:
        img = cv2.imread(f"seg_dataset_saranya2021/Yellowish_Green/{str(i)}.jpg")
    if i%4 ==2:
        img = cv2.imread(f"seg_dataset_saranya2021/Midripen/{str(i)}.jpg")
    if i%4 ==3:
        img = cv2.imread(f"seg_dataset_saranya2021/Overripen/{str(i)}.jpg")
    rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # BGR to RGB 
    plt.imshow(rgb_img)
    plt.axis('off')
    plt.show()
```

```python
flags_yes = np.array([0,1,1,0,0,1,0,0])
```

```python
def get_banana_preference(flags_for_like):
    res = 0.0
    for i in range(flags_for_like.shape[0]):
        res=res+(i%4)*flags_for_like[i]
    return res/np.sum(flags_for_like)
```

```python
print(get_banana_preference(flags_yes))
```
